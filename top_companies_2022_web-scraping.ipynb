```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"
r= requests.get(url)
```

```python
print(r.content,"html5lib")
```

```python
soup= BeautifulSoup(r.text)
```

```python
print(soup)
```

```python
table=soup.find_all("table")[0]
```

```python
print(table)
```

```python
title=table.find_all("th")
```

```python
print(title)
```

```python
title=[title.text.strip() for title in title]
```

```python
print(title)
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(columns=title)
```

```python
df
```

```python
columns_data=table.find_all("tr")
```

```python
for row in columns_data[1:]:
    row_data=row.find_all("td")
    individual_data=[data.text.strip() for data in row_data]
    # print(individual_data)

    import numpy as np

    a=[]
    a=individual_data

    a.pop (1)

    a.pop (2)

    length= len(df)
    df.loc[length]=a
```

```python
df
```

```python
df.to_csv("Top_company_2022.csv", index=False)
```
